```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=500, n_features=1, noise=15, random_state=4)
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

```python
plt.scatter(X, y, label='Data points')
# Add labels and title
plt.xlabel('X')
plt.ylabel('y')
plt.title('Plot of Synthetic Data')
# Show the plot
plt.show()
```

```python
class LinearRegression1():
    def __init__(self, lr: int = 0.01, n_iters: int = 1000) -> None:
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.weights = np.random.rand(num_features)
        self.bias = 0

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

```python
model = LinearRegression1()
model.fit(X,y)
preds = model.predict(X)
```

```python
# Plot the data points
plt.scatter(X, y, label='Data points')

# Plot the predictions
plt.plot(X, preds, color='black', label='Predictions')

# Add labels and title
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Fit')

plt.show()
```
