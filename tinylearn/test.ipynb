```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression as skLR

from tiny.linear_regression import LinearRegression
```

# Data

```python
X, y = make_regression(n_samples=500, n_features=1, noise=15, random_state=4)
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")

plt.scatter(X, y, label='Data points')
# Add labels and title
plt.xlabel('X')
plt.ylabel('y')
plt.title('Plot of Synthetic Data')
# Show the plot
plt.show()
```

# Our implementation

```python
model = LinearRegression(n_iters=500)
model.fit(X, y)
preds = model.predict(X)

# Plot the data points
plt.scatter(X, y, label='Data points')

# Plot the predictions
plt.plot(X, preds, color='black', label='Predictions')

# Add labels and title
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Fit')

plt.show()
```

# Sklearn implementation

```python
model_sk = skLR()
model_sk.fit(X,y)
preds_sk = model_sk.predict(X)

# Plot the data points
plt.scatter(X, y, label='Data points')

# Plot the predictions
plt.plot(X, preds_sk, color='black', label='Predictions')

# Add labels and title
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Fit')

plt.show()
```

# Mean squared errors

```python
print(f"Ours implementations MSE is: {mean_squared_error(preds, y):.2f}")
print(f"Sklearn implementations MSE is: {mean_squared_error(preds_sk, y):.2f}")
```
